This case gives scalars their name, the scalar "scales their vector
w = λ**v** 

The scalar doesnt official change the 'direction' or angle of **v**

There are several cases here
## λ>1 
the resulting vector will point in the same direction but be greater than the original vector

## λε(0,1)
When the scalar is between 0 and 1 the resulting vector will be smaller

## λ<0 
When the scalar is less than 0 (negative) the resulting vector, 'points in another direction', this isnt accurate but it's an appropriate mental model for now
